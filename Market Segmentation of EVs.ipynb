```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**DATA PRE-PROCESSING**

```python
df1 = pd.read_csv('/content/EVIndia.csv')
df1
```

```python
# dropping unwanted columns
df2 = df1.drop(['BaseModel', 'TopModel'], axis=1)
df2
```

```python
# replacing na -> 0
df3 = df2.replace('na' , 0 , regex = True)
df3
```

```python
# removing UNITS
df3['Range'] = df3['Range'].replace('Km/Full Charge' , '' , regex  = True)
df3['BootSpace'] = df3['BootSpace'].replace('L' , '' , regex  = True)
df3['Capacity'] = df3['Capacity'].replace('Seater','',regex = True)
```

```python
# changing datatype
df3['Range'] = df3['Range'].astype("int64")
df3['BootSpace'] = df3['BootSpace'].astype("int64")
df3['Capacity'] = df3['Capacity'].astype("int64")
```

```python
# replacing values of 0 -> mean of that column
df3['Range'] = df3['Range'].replace([0],[df3['Range'].mean()])
df3['BootSpace'] = df3['BootSpace'].replace([0],[df3['BootSpace'].mean()])
df3
```

```python
df3 = df3.replace(0 , 'Hyundai Kona Electric' , regex = True)
df3
```

```python
# manually gathered BATTERY CAPACITY (KWH) of above mentioned EVs from Google
b_c = [30.5,26,40.5,50.3,39.2,90,93.4,71.7,80,71,93.4,95]
df3.insert(5, 'BatteryCap',b_c, False)
df3
```

```python
# changing PRICE RANGE column to a SINGLE PRICE UNIT (in lakhs)
pr = [17.4,13.64,19.24,25.88,23.98,112,180,29.15,100,116,150,119]
df3.insert(6, 'Price', pr, False)
df3
```

```python
# DROPPPING EXCESSIVE PRICE RANGE COLUMN
df4 = df3.drop(['PriceRange'], axis=1)
df4
```

```python
# manually gathered TOP SPEED (kmph) of above mentioned EVs from Google
t_s = [100,80,120,140,167,200,245,160,180,200,260,200]
df4.insert(6, 'TopSpeed', t_s, False)
df4
```

```python
# input
x = df4.iloc[:, [6,7]].values
```

```python
# K-MEANS CLUSTERING
# elbow method to find the best number of clusters for this data
from sklearn.cluster import KMeans 
k = range(1,10) 
sse = []
for i in k:
  model_demo = KMeans(n_clusters=i,random_state = 0)
  model_demo.fit(x)                                     #we get the sse value 
  sse.append(model_demo.inertia_) 
plt.figure(figsize = (15,7))
plt.scatter(k,sse)
plt.plot(k,sse)
plt.show()
```

```python
# try , k = 5
k = 5 
from sklearn.cluster import KMeans 
model = KMeans(n_clusters=k,random_state = 0)
model.fit(x)
y = model.predict(x)
```

```python
plt.scatter(x[y == 0, 0], x[y == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(x[y == 1, 0], x[y == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(x[y == 2, 0], x[y == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(x[y == 3, 0], x[y == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(x[y == 4, 0], x[y == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')

plt.title('Clusters of customers')
plt.xlabel('TopSpeed')
plt.ylabel('Price')
plt.legend()
plt.show()
```
